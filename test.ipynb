```python
from services.preprocessing import Preprocessing
import json
```

```python
data_path = r"D:\College\Research\Codes\Dataset\tweets-data-ruu.json"
with open(data_path, "r", encoding="utf-8") as f:
    raw_doc = json.load(f)
docs = [
    str(entry["full_text"]) if isinstance(entry.get("full_text"), str)
    else ""  # Convert NaN or dicts to empty string
    for entry in raw_doc
]
id_str_list = [
    # plain integer → stringify
    str(entry["id_str"])
    if isinstance(entry.get("id_str"), int)

    # nested {"$numberLong": "..."} → pull out the string
    else entry["id_str"]["$numberLong"]
    if isinstance(entry.get("id_str"), dict)
       and "$numberLong" in entry["id_str"]

    # everything else → empty string
    else ""
    for entry in raw_doc
]
```

```python
docs = docs[:10]
id_str_list = id_str_list[:10]
```

```python
tweet = Preprocessing(docs, id_str_list, "ruu tni").get_data()
```
